```python
#Unique Words in a Sentence
sentence = "python is easy and python is powerful"
words = sentence.split()
unique_words = set(words)
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

```python
#Highest Salary from Employee Data
employees = {"Ravi": 75000, "Anita": 68000, "Kiran": 72000}
max_salary = 0
top_employee = ""
for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        top_employee = name
print(f"Highest Salary: {top_employee} - {max_salary}")
```

```python
#Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
max_num = numbers[0]
min_num = numbers[0]
for n in numbers:
    if n > max_num:
        max_num = n
    if n < min_num:
        min_num = n
print(f"Max: {max_num}")
print(f"Min: {min_num}")
```

```python
#Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
expensive_count = 0
for p in prices:
    if p > 1000:
        expensive_count += 1
print(f"Products above 1000: {expensive_count}")
```

```python
#Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
presents = attendance.count("P")
total_days = len(attendance)
score = (presents / total_days) * 100
print(f"Attendance Percentage: {score}")
```

```python
#Remove Duplicate Phone Numbers
phones = [9876543210, 9123456789, 9876543210]
unique_phones = set(phones)
print(f"Unique phone numbers: {unique_phones}")
```

```python
#Count Character Frequency
word = "pythonp"
counts = {}
for char in word:
    if char in counts:
        counts[char] += 1
    else:
        counts[char] = 1

print(counts)
```
