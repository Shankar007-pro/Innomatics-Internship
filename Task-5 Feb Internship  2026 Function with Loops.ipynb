```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
def check_rainfall_adequacy(rainfall_data, required_level):
    # Step 1: Calculate the total using a loop
    total_rainfall = 0
    for amount in rainfall_data:
        total_rainfall += amount
    
    # Step 2: Calculate the average
    # We divide the total by the number of items in the list
    average_rainfall = total_rainfall / len(rainfall_data)
    
    # Step 3: Determine status using conditional logic
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # Step 4: Print the expected format
    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}")

# --- Real-World Test Case ---
# Seasonal data for 4 months and a required average of 70
seasonal_data = [80, 65, 70, 73]
required_threshold = 70

check_rainfall_adequacy(seasonal_data, required_threshold)
```

```python
#Problem 4: Social Media – Duplicate Account Detection
def detect_duplicate_accounts(usernames):
    # Step 1: Count how many usernames we started with
    original_count = len(usernames)
    
    # Step 2: Convert to a set (this automatically removes duplicates)
    unique_usernames = set(usernames)
    unique_count = len(unique_usernames)
    
    # Step 3: Compare lengths to determine status
    if unique_count < original_count:
        status = "Yes"
    else:
        status = "No"
    
    # Step 4: Output the result in the expected format
    print(f"Duplicate Accounts Found: {status}")

# --- Real-Time Test Cases ---
# Case A: Has duplicates ("user_alpha" appears twice)
social_media_users = ["user_alpha", "coder_123", "user_alpha", "tech_guru"]
detect_duplicate_accounts(social_media_users)

# Case B: All unique
new_platform_users = ["admin_01", "mod_team", "guest_user"]
detect_duplicate_accounts(new_platform_users)
```

```python
#Problem 5: Healthcare – Appointment Eligibility Checker

def check_appointment_eligibility(age):
    # Step 1: Apply the eligibility rule using conditional logic
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    # Step 2: Print the results in the required format
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

# --- Real-Time Test Cases ---
# Testing with an adult patient
check_appointment_eligibility(21)

print("-" * 20) # Just a separator

# Testing with a minor patient
check_appointment_eligibility(16)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(crop_prices):
    # Step 1: Define the premium threshold
    threshold = 2000
    
    # Step 2: Use list comprehension to filter prices
    # This reads: "Keep price for every price in crop_prices IF price is > 2000"
    premium_crops = [price for price in crop_prices if price > threshold]
    
    # Step 3: Print the result in the expected format
    print(f"Premium Crops: {premium_crops}")

# --- Real-World Test Case ---
# A list of prices from the morning market
current_market_prices = [1500, 2500, 1800, 3200, 2000, 4500]

filter_premium_crops(current_market_prices)
```

```python
#Problem 7: System Monitoring – Application Health 
def check_app_health(error_count):
    # Step 1: Use conditional statements to categorize health
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    # Step 2: Print the results in the required format
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# --- Real-Time Test Cases ---
# Scenario: System is currently experiencing high load
check_app_health(7)

print("-" * 20)

# Scenario: System is stable
check_app_health(0)
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker
def validate_transaction(amount):
    # Step 1: Define the daily limit
    daily_limit = 50000
    
    # Step 2: Use conditional logic to check the amount
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    # Step 3: Print the results in the required format
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# --- Real-Time Test Case ---
# A customer tries to transfer ₹60,000
validate_transaction(60000)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance_list):
    # Step 1: Use a loop to count total days present
    # (Though sum(attendance_list) is faster, we'll use a loop per the hint)
    total_present = 0
    for day in attendance_list:
        if day == 1:
            total_present += 1
            
    # Step 2: Calculate the percentage
    total_days = len(attendance_list)
    attendance_percentage = (total_present / total_days) * 100
    
    # Step 3: Determine eligibility based on the 75% rule
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
        
    # Step 4: Output the results in the expected format
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {eligibility}")

# --- Real-Time Test Case ---
# 1 = Present, 0 = Absent (Student attended 4 out of 5 days)
student_attendance = [1, 1, 1, 0, 1] 

check_exam_eligibility(student_attendance)
```

```python
#Problem 10: Smart Electricity Bill Analyzer
def analyze_electricity_bill(units):
    total_bill = 0
    
    # Step 1: Calculate the bill based on slabs
    if units <= 100:
        # Slab 1: 1-100 units
        total_bill = units * 3
    elif units <= 200:
        # Slab 1 (First 100) + Remaining in Slab 2
        total_bill = (100 * 3) + (units - 100) * 5
    else:
        # Slab 1 (First 100) + Slab 2 (Next 100) + Remaining in Slab 3
        total_bill = (100 * 3) + (100 * 5) + (units - 200) * 7
        
    # Step 2: Classify usage level based on the total bill
    if total_bill < 500:
        usage_level = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_level = "Moderate Usage"
    else:
        usage_level = "High Usage"
        
    # Return both values as requested in the hint
    return total_bill, usage_level

# --- Real-Time Execution ---
consumed_units = 250
bill_amount, status = analyze_electricity_bill(consumed_units)

print(f"Total Bill: ₹{bill_amount}")
print(f"Usage Level: {status}")
```
