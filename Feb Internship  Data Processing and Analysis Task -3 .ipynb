```python
#Problem 1: Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# 1. Identify the highest performance score
max_score = max(employees.values())

# 2. Handle ties by finding all employees with the max_score
top_performers = [name for name, score in employees.items() if score == max_score]

# 3. Display results
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
#Search Query Keyword Analysis
query = "Buy mobile phone buy phone online"

# 1. Convert to lowercase and split into words
words = query.lower().split()

# 2. Count frequency of each keyword
word_counts = {}
for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1

# 3. Filter and display only keywords searched more than once
duplicates = {word: count for word, count in word_counts.items() if count > 1}
print(duplicates)
```

```python
#Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# 1. Identify even readings and store as (hour, value) tuples
# Hour index starts at 0, but usually corresponds to the list index here
valid_readings = [(index, val) for index, val in enumerate(sensor_readings) if val % 2 == 0]

# 2. Display output
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# 1. Extract domains and count them
domain_counts = {}
for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

# 2. Calculate percentage and display
total_emails = len(emails)
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

# 1. Calculate average daily sales
avg_sales = sum(sales) / len(sales)

# 2. Detect spikes (> 30% above average)
# We use index + 1 to show 'Day 1' instead of 'Day 0'
for i, sale in enumerate(sales):
    if sale > (avg_sales * 1.3):
        print(f"Day {i + 1}: {sale}")
```

```python
#Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# 1. Count occurrences
id_counts = {}
for uid in user_ids:
    id_counts[uid] = id_counts.get(uid, 0) + 1

# 2. Identify and display duplicates
for uid, count in id_counts.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
