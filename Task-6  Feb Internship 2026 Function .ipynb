```python
#1st problem
def manage_parking_lot(capacity, logs):
    current_vehicles = 0
    peak_usage = 0

    for action in logs:
        if action == "IN":
            current_vehicles += 1
            # Update peak usage if current count exceeds the previous peak
            if current_vehicles > peak_usage:
                peak_usage = current_vehicles
        
        elif action == "OUT":
            # Safety check: Parking count shouldn't go below 0
            if current_vehicles > 0:
                current_vehicles -= 1

        # Real-time Capacity Alert
        if current_vehicles >= capacity:
            print(f"ALERT: Parking Lot is FULL! ({current_vehicles}/{capacity})")

    # Determine final status
    status = "Full" if current_vehicles >= capacity else "Available"
    
    return current_vehicles, peak_usage, status

# --- Testing the Function ---
capacity_limit = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

current, peak, status = manage_parking_lot(capacity_limit, vehicle_logs)

print(f"Currently Parked Vehicles: {current}")
print(f"Peak Usage Today: {peak}")
print(f"Parking Status: {status}")
    
    
```

```python
#2nd problem
def estimate_delivery(distance, weather, traffic):
    # Base calculation: 5 minutes per km
    base_time = distance * 5
    
    # Apply Traffic Delays
    traffic_delays = {"Low": 0, "Medium": 10, "High": 20}
    
    # Apply Weather Delays
    weather_delays = {"Clear": 0, "Rainy": 15, "Stormy": 30}
    
    total_eta = base_time + traffic_delays.get(traffic, 0) + weather_delays.get(weather, 0)
    
    print(f"Estimated Delivery Time: {total_eta} minutes")
    return total_eta

# Sample Input
estimate_delivery(8, "Rainy", "High")
```
