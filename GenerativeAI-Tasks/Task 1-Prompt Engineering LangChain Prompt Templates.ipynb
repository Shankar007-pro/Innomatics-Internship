```python
# Task 1 & 2: Multi-Input Prompt System
try:
    from langchain_core.prompts import PromptTemplate
except ImportError:
    from langchain.prompts import PromptTemplate

template_text = "Explain {topic} for {audience} in a {tone} tone."

# Creating the template object
# input_variables is key here to avoid hardcoding
prompt_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template_text
)

# Generating the dynamic prompt
output = prompt_template.format(topic="AI", audience="beginners", tone="friendly")
print(output)
```

```python
# Task 3: Prompt Variations Engine
# Define the templates
teaching_template = PromptTemplate(input_variables=["topic"], template="Teaching -> Explain {topic} clearly step by step")
interview_template = PromptTemplate(input_variables=["topic"], template="Interview -> Ask 3 questions about {topic}")
story_template = PromptTemplate(input_variables=["topic"], template="Storytelling -> Explain {topic} as a story")

# Run for topic = Machine Learning
target_topic = "Machine Learning"

print(teaching_template.format(topic=target_topic))
print(interview_template.format(topic=target_topic))
print(story_template.format(topic=target_topic))
```

```python
from langchain_core.prompts import ChatPromptTemplate

# Create the Chat system
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a professional {role}."),
    ("human", "Explain {topic} to me.")
])

# Test Case: role = teacher, topic = Neural Networks
print("--- Chat Prompt Output ---")
formatted_chat = chat_template.format_messages(role="teacher", topic="Neural Networks")

for message in formatted_chat:
    print(f"{message.type.upper()}: {message.content}")
```

```python
# Task 5: Input Validation Layer
def validate_inputs(audience, tone):
    allowed_audiences = ['beginner', 'intermediate', 'expert']
    allowed_tones = ['formal', 'casual', 'fun']
    
    # Validation Logic
    if audience not in allowed_audiences:
        print(f"⚠️ Invalid audience '{audience}'. Defaulting to 'beginner'.")
        audience = 'beginner'
        
    if tone not in allowed_tones:
        print(f"⚠️ Invalid tone '{tone}'. Defaulting to 'formal'.")
        tone = 'formal'
        
    return audience, tone

# Quick Test
print(validate_inputs("kid", "silly")) # Should default
```

```python
# Task 6: Prompt Generator App
def generate_prompt(topic, audience, tone, style):
    # 1. Validate inputs using our Task 5 function
    v_audience, v_tone = validate_inputs(audience, tone)
    
    # 2. Define style-based templates
    styles = {
        "teaching": "Explain {topic} clearly step by step for {audience} in a {tone} tone.",
        "interview": "Ask 3 interview questions about {topic} for an {audience} in a {tone} tone.",
        "storytelling": "Explain {topic} as a story for {audience} in a {tone} tone."
    }
    
    # Get template or default to teaching
    template_str = styles.get(style.lower(), styles["teaching"])
    
    # 3. Create LangChain PromptTemplate (No f-strings in template!)
    prompt_obj = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template_str
    )
    
    return prompt_obj.format(topic=topic, audience=v_audience, tone=v_tone)

# Example Output for Task 6 requirements
print("--- Task 6 Result ---")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
# Task 7: Reusability Test
# We use ONE single template structure
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Guide for {topic}: Targeted at {audience} with a {tone} approach."
)

# 5 Different Inputs
test_inputs = [
    {"topic": "Cloud Computing", "audience": "expert", "tone": "formal"},
    {"topic": "Cooking", "audience": "beginner", "tone": "casual"},
    {"topic": "Stock Market", "audience": "intermediate", "tone": "formal"},
    {"topic": "Cybersecurity", "audience": "beginner", "tone": "fun"},
    {"topic": "Yoga", "audience": "expert", "tone": "casual"}
]

print("--- Task 7: Running 5 Inputs through 1 Template ---")
for data in test_inputs:
    print(reusable_template.format(**data))
```
