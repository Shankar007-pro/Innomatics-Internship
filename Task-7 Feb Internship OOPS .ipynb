```python
# problem 1
class SmartLight:
    def __init__(self, name):
        self.name = name
    def set_status(self, action):
        print(f"{self.name} is {action}")
light_name = "Bedroom Light"
action = "ON"
light = SmartLight(light_name)
light.set_status(action)
```

```python
#problem 2
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
emp = Employee("Rahul", "EMP102", "IT")
emp.display_card()
```

```python
#problem 3
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
contact = Contact("Anita", "9876543210")
contact.display_contact()
```

```python
#problem 4
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def print_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
item = Product("Headphones", 2499)
item.print_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")
film = Movie("Inception", 4.8)
film.display_details()
```

```python
class DeliveryManager:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")
delivery = DeliveryManager("Suman", "Hyderabad")
delivery.print_details()
```
