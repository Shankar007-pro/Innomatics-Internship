```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        try:
            amount = int(input("Enter recharge amount: "))
            
            if amount < 50:
                print("Minimum recharge amount is ₹50. Please try again.")
                continue
            
            if amount in valid_plans:
                print(f"Recharge of ₹{amount} successful!")
                break
            else:
                print(f"Invalid plan. Available plans are: {valid_plans}")
                retry = input("Try again? (yes/no): ").lower()
                if retry != 'yes': break
        except ValueError:
            print("Invalid input. Please enter a numerical value.")

validate_recharge()
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"ALERT: {product} is low on stock ({quantity}). Reorder Alert!")
        else:
            print(f"Stock OK: {product} ({quantity})")

# Example Data
inventory = {"Apples": 20, "Bananas": 10, "Oranges": 14, "Grapes": 50}
check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System
def process_student_result(marks_list):
    if not marks_list: return "No marks provided"
    
    total = 0
    for mark in marks_list:
        total += mark
    
    average = total / len(marks_list)
    
    if average >= 50:
        return f"Average: {average:.2f} - Status: Pass"
    else:
        return f"Average: {average:.2f} - Status: Fail"

# Usage
marks = [55, 60, 45, 70, 30]
print(process_student_result(marks))
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(km, is_peak):
    base_fare = 50
    rate_per_km = 12
    total = base_fare + (km * rate_per_km)
    
    if is_peak:
        total *= 1.25  # 25% extra
    
    return total

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower() == "yes"
    
    fare = calculate_fare(distance, peak)
    print(f"Your estimated fare is: ₹{fare:.2f}")
    
    if input("Calculate another fare? (yes/no): ").lower() != "yes":
        break
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def check_attendance_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    
    for status in attendance_list:
        if status == "P":
            present_days += 1
            
    percentage = (present_days / total_days) * 100
    
    if percentage >= 75:
        return f"Percentage: {percentage}% - Eligible"
    else:
        return f"Percentage: {percentage}% - Not Eligible"

# Usage
attendance = ["P", "P", "A", "P", "P", "A", "P"]
print(check_attendance_eligibility(attendance))
```

```python
#Problem 6: Password Strength Checker
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Minimum 8 characters required."
    
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
            
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak: Must contain at least one digit and one special character (@#$)."

# Usage
pwd = input("Enter password: ")
print(check_password_strength(pwd))
```
